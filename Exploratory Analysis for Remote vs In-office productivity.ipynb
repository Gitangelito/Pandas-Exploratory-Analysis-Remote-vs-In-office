```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Employment_Type': ['In-Office', 'Remote'],
        'Count': [510, 490]}

# Create a DataFrame
df = pd.DataFrame(data)

# Count the occurrences of each employment type
employment_type_counts = df['Employment_Type'].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(df['Count'], labels=df['Employment_Type'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Employment Types')
plt.axis('equal')  # Equal aspect ratio ensures a circular pie chart
plt.show()
```

```python
#Analizing relationships:

import pandas as pd
import seaborn as sns

sns.pairplot(df)
plt.show()
```

```python
import pandas as pd
import numpy as np

# Select numerical columns
numerical_columns = df.select_dtypes(include=[np.number])

# Calculate correlation matrix
correlation_matrix = numerical_columns.corr()

# Create heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```
